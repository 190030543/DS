```python
import pandas as pd
import seaborn as sns
```

```python
#Give names to the columns of the above dataset
df = pd.read_csv("pima-indians-diabetes.csv",names = ["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","HasDiabetes"])
df
```

```python
#Print first 2 records
df.head(2)
```

```python
#Check if there are any columns with empty/null dataset?
df.isna().sum()
```

```python
#Find correlation between all predative features w.r.t "HasDiabetes" (target) feature.
df.corrwith(df["HasDiabetes"])
```

```python
# See carefully and find 3 most correlated features to target variable.
df.corrwith(df["HasDiabetes"]).sort_values(ascending=False)[1:4]
```

```python
#Plot a heatmap of the correlation matrix obtained above.
sns.heatmap(df.corr(),annot=True)
```

```python
#Calculate the median value for BMI and substitute it in the BMI column where its value is 0 
df["BMI"] = df["BMI"].replace(0,df["BMI"].median())
df
```

```python
# Calculate the median value for each of them and substitute it in the respective column in place of 0
df["Pregnancies"] = df["Pregnancies"].replace(0,df["Pregnancies"].median())

df["Glucose"] = df["Glucose"].replace(0,df["Glucose"].median())

df["SkinThickness"] = df["SkinThickness"].replace(0,df["SkinThickness"].median())

df["Insulin"] = df["Insulin"].replace(0,df["Insulin"].median())   

df["DiabetesPedigreeFunction"] = df["DiabetesPedigreeFunction"].replace(0,df["DiabetesPedigreeFunction"].median())

df["Age"] = df["Age"].replace(0,df["Age"].median())

df
```
